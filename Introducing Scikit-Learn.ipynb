# Scikit-Learn's Estimator API(application programing interface)
Steps in using Scikit Learn estimator API by using Simple Linear Regresion

```python
import numpy as np
import matplotlib.pyplot as plt

rng=np.random.RandomState(42)
x=10*rng.rand(50)
y=3*x+1-rng.randn(50)

plt.scatter(x,y);
```

```python
# Step1:- Choose a class of model

from sklearn.linear_model import LinearRegression  

# Here LinearRegression is simple Linear Regression model class in linear_model class
```

```python
# Step2:- Choose model hyperparameters

model=LinearRegression(fit_intercept=True)
model

# fit_intercept is one of the hyperparameter
```

```python
# Step 3:- Arrange data into feature matrix and target vector

X=x[:,np.newaxis]
X.shape

# np.newaxis adds one extra dimension of an existing array
```

```python
# Fit model to our data

model.fit(X,y)

print(model.coef_)  # model parametersthat were learned during model fitting

print(model.intercept_)
```

```python
# Predict labels for unlabel data

xfit=np.linspace(-1,11) # 'new data'
ytest=3*x+1-xfit


Xfit=xfit[:,np.newaxis]
yfit=model.predict(Xfit)

plt.scatter(x,y)
plt.plot(xfit,yfit);
```

```python
# Error calculation

from sklearn import metrics

print('RMSE:', np.sqrt(metrics.mean_squared_error(ytest, yfit)))
```
